## Machine Learning(ML) Basic Concepts

The algorithms behind machine learning are very complex mathematics. We may never understand those if math is not our forte. So, let's define some basic concepts without the complexities of mathematics to get us going.

**What is Machine Learning?**

In the simplest description, it is a science of creating a program to learn from existing data and be able to predict a required output when fed by a new data. It is very exciting field and it is the driving force behind the widespread application of Artificial Intelligence(AI) as we know today.


**Artificial Neural Network**

<img src="images/ann.png" style="width: 600px">

Image: CC BY 1.0

There are a lot of algorithm for machine learning. One popular one is by using Artificial Neural Networks. This is the *brain* of machine learning. It is inspired by how the *"neurons"* in our brain works. It consists of the *Input Layer*, *Hidden Layer* and an *Output Layer*. During our learning/training process, this is commonly referred to as model. But technically the model is the result of the process.

In our image, the *input layer* consists of two neurons denoted by the green circles. The number of neurons must be equal to the number of features of the dataset being used.

The *hidden layer* consists of the five cyan circles, is arbitrary. These arbitrary parameters are commonly called hyperparameters. Determining the number of layers and the number of neurons is the developer's decision. This is how the term *training* in machine learning came up. We have to *train* over and over our *machine* with different number of hidden layers and neurons to *learn* from our input data. Try 1 hidden layer with 64 neurons first. Then start adding or removing neurons in $n^2$. Thus, you can try 8, 16, 32, 64, 128, 256 and so on. 10, 50, 100 neurons can be used perfectly. But because binary system has been used since the dawn of computers, it seems logical to choose the $n^2$ values. If you decide to use more than one hidden layer, it is advised to use the same number of neurons for each hidden layer. Usually, adding more than one hidden layer is better. 

The *output layer*, denoted by the yellow circle, must be equal to the number of predicted label or result. Example, if we are trying to predict the (1)price of a house, the number of neurons would be 1. Likewise, if our neural network is trying to classify if a review is (1)good or (2)bad, our neurons would be 2. If, on the other hand, we are classifying a news if it is (1)political, (2)health, (3)entertainment or (4)sports news, then our neurons would be 4.

**Loss**

This the summation of errors during training and validation. It is safe to say that the lower the loss, the better our model.

**Bias**

This is an extra neuron with a value of 1 added to the input of a layer in an artificial neural network. This is added so that it can be able to *train* our model better. It is best discussed technically [here](https://stackoverflow.com/questions/2480650/role-of-bias-in-neural-networks). If at the beginning you don't understand this, don't worry. As soon as you do more and more machine learning problems, you will understand these concepts in due time.

**Accuracy**

This is the percentage of the model's accuracy in predicting the desired output. So, if we have 100 test samples and our model was able to predict 95 correctly (in the case of classifying if a review is good or bad), our accuracy is 95.00%

**Convergence**

Technically, we can say that our training has converge when training has reached the minimum loss.

**Supervised vs Unsupervised Learning**

Supervised learning means that we have to "teach" our machine to learn something. While an unsupervised learning means that the machine will "learn" by itself. An easy way to determine if we will using supervised or unsupervised learning is to check the dataset given. If the dataset is labeled, like in our example, the prices of the house are given and we are trying to predict the price for new house, it is a supervised learning. If the dataset is not labeled and we are deriving some relationships in our dataset, this is unsupervised learning. An example is a dataset of customer purchases and we would like to do customer segmentation based on purchased products.

There are other kinds of learning like reinforcement learning, but we will focus on these two most popular learning.

**Machine Learning Problems**

The two most common machine problems are:
- Regression - predict continous value (e.g. the price of a house)
- Classification - predict discrete value (e.g. 0=Bad review, 1=Good review)

There are different types of neural networks, too, like Convolutional Neural Network(CNN), Recurrent Neural Network(RNN), Long Short-Term Memory(LSTM) and others but first, let's build our skill on the basic neural network presented earlier.

## The goal

Our goal is to predict the MPG of a car. We will use the [Auto MPG](https://archive.ics.uci.edu/ml/datasets/Auto+MPG) from UCI. From the given dataset, we will consider this as a supervised learning and it is a regression problem.

## Learning by practical coding

Let's begin.

<div class="alert alert-block alert-info">
<b>Import all the modules we need</b>
</div>
We need some supporting modules in order to process the data, displaying visual aids and other functions. If the module is not available in your python environment, you can install it via pip.

```python
import keras #the keras high-level API
import pandas as pd #for manipulating dataset
import seaborn as sns #plotting data
import matplotlib.pyplot as plt #plotting data
import sklearn.metrics as metrics #for computing metrics after the training
from sklearn.model_selection import train_test_split #function to split our data
from keras.models import Sequential #neural network model
from keras.layers import Dense #neural network layers
from keras.optimizers import RMSprop #optimizer for linear regression
from keras.callbacks import ModelCheckpoint, EarlyStopping #callbacks during training
```

<div class="alert alert-block alert-info">
<b>Read the dataset</b>
</div>

```python
#read the data from file using pandas
#but first let's define the name of the columns
#as described in auto-mpg.names

filename = "./dataset/auto-mpg.data"
column_names = ["mpg", "cylinders", "displacement", "horsepower", "weight", 
                "acceleration", "model year","origin"]

df = pd.read_csv(filename, names=column_names,
                 sep= " ",#the columns are separated by spaces
                 skipinitialspace=True, #so let's skip the other spaces between the columns
                 na_values="?",#missing values in the raw data is indicated by character ?
                 comment="\t"#each row has a column car name which is separated by tab. ignore this column
                 )

#display the top 5 rows to check if we are reading it correctly
df.head()
```

<div class="alert alert-block alert-info">
<b>Check for missing values</b>
</div>

```python
#check for missing values using built-in function in pandas
df.isna().sum()
#we can see that there are 6 missing values in the horsepower column
```

<div class="alert alert-block alert-info">
<b>Clean dataset</b>
</div>

```python
#there are some techniques like imputation to fill missing values
#but for this tutorial let's just drop this rows
df.dropna(inplace=True)

#let's check if they missing values are deleted
df.isna().sum()
#and they are now gone
```

<div class="alert alert-block alert-info">
<b>One-hot encode categorical features</b>
</div>
It is a process of converting categorical variables into numerical form. We also have to create additional features equivalent to all the categories. Categorical features should be one-hot encoded to get a better model. It can be best explained by looking before and after encoding.

**Dataset before one-hot encoding**

```python
#the column origin is a categorical feature so let's encode it
#To understand this, let's display the dataset before one hot encode.
#dataset before one hot encoding
df.head()
```

**Dataset after one-hot encoding**

```python
#pandas has built-in one hot encoder
#let's use it to encode the column "origin"
df_encoded = pd.get_dummies(df, columns=["origin"])

#display first 5 rows
df_encoded.head()
```

As you can see, the *origin* column is removed and replaced with three columns. There are three columns because there are three categories for this column.

<div class="alert alert-block alert-info">
<b>Derive new features</b>
</div>
There are certain circumstances that we will derive new features based on the existing one. This would be as simple as adding two columns or as complex as creating another model. We don't need one for our current dataset.

<div class="alert alert-block alert-info">
<b>Visualize the data</b>
</div>
A picture says a thousand words. To better understand our data, a visual representation is the way to go. It may also help us evaluate which features to use.

```python
#use seaborn to plot our data
#a very simple way to plot
sns.pairplot(df)
```

You can see that some plots follow a linear equation like horsepower vs displacement.

<div class="alert alert-block alert-info">
<b>Select features</b>
</div>
    
It is very important to select the right features to come up with an accurate model. These are hyperparameters and during our training we can add or remove some to improve accuracy. Let's select them all for now.

<div class="alert alert-block alert-info">
<b>Create the training and test data.</b>
</div>

A good practice is to train our model with a training data and test separately with a test data. We should **never** use the training data to test our model. If we test our model with the training data, obviously, it gives us a good result. But what we are after is how well it will predict unseen data which is our test data that was never used in the training.


```python
#it is a best practice to shuffle the dataset
#so that we do not end up training to a just speficic set of data specially if the dataset is sorted
#the function automatically randomize the dataset
#but we have to set the random_state to a fixed integer 
#so that we can reproduce with the same arrangement of dataset every time
#test_size means that we will create a 20% test data which is a good ratio to start with
df_train, df_test = train_test_split(df_encoded, test_size=0.2, random_state=0)

#get the training label
y_train = df_train["mpg"]

#and the test label
y_test = df_test["mpg"]

#remove the label from train and test data
df_train.pop("mpg")
df_test.pop("mpg")

#display top 5
df_train.head()
#you may have noticed that the index, the left most column is already shuffled
```

<div class="alert alert-block alert-info">
<b>Normalize the data</b>
</div>

Normalization is the process of transforming all the values of our features to as close as possible. We can train our model without normalization but learning may take a very long time or if we are in bad luck, our model may not converge at all.

A good rule of thumb is if difference between features is in hundreds, it is better to normalize the data. Especially for big dataset, we should normalize all features in the range -1..1.

It should be noted that the training data and test data should be normalized with the same technique.

There are many techniques for normalization. One of my favorites is by subtracting the minimum value then dividing it with the maximum value.

```python
#before normalization let's get the statistics of the dataset to determine the min/max value of each column
df_stat = df_train.describe()

#let's transpose the statistics.
#that is writing rows(count, mean, max, etc) as columns and vice versa
df_stat = df_stat.transpose()

#display the statistics
df_stat
```

```python
#define a function to do the normalization
def normalize(dataset, stats):
    '''Returns a normalized dataframe.'''
    return (dataset - stats["mean"])/ stats["max"]
   
#normalize the training and test data to temporary objects
#we are doing this so that our original dataset will not be messed up
#in case we are restarting the execution of the program at this point
x_train_temp_norm = normalize(df_train, df_stat)
x_test_temp_norm = normalize(df_test, df_stat)

#let's make a copy of our temporary train and test data
x_train =  x_train_temp_norm.copy()
x_test = x_test_temp_norm.copy()

#display top 5 rows
x_train.head(5)
```

<div class="alert alert-block alert-info">
<b>Create a model.</b>
</div>
Now that we are finished processing our dataset. We can not start machine learning.

```python
#this is how to create the model of our neural network for a regression problem
#keras Sequential model is a linear stack of layers and it is the model that works most of the time
#you must start by creating a model first before stacking it with layers
model = Sequential()

#add input layer
#Dense is just a regular densely-connected neural layer
#units = the number of neurons
#activation = defines an output signal that can be fed to the input signal of the next layer
#it "connects" one layer to another
#relu = Rectified Learning Unit
#there are other types of activation but this is the most widely used for hidden layers
#relu cannot be used in the output layer
#adding the first layer actually adds a hidden layer with neurons equal to number specified in units
#in this case 64
#the input_dim parameter creates the input layer with neurons
#this is the number of features of our training data
INPUT_DIM = x_train.shape[1]
model.add(Dense(units=64, input_dim=INPUT_DIM, activation="relu"))

#add another hidden layer with the same neurons as the previous one
model.add(Dense(units=64,  activation="relu"))

#add output layer
#there is only 1 unit here because our output is only one, that is the price of a house
#activation is set to linear
#for problems with predicting a single real value (regression), a linear activation is used
#it just outputs whatever the values from the previous layer lineary
#we no longer specify the input_dim because it is implicit
#and keras will automatically assign it based from previous layer
model.add(Dense(units=1, activation="linear"))

#after adding all the layers, let's compile the model.
model.compile(
              #specify the type of loss to check, that is Mean Squared Error
              loss="mse",          
              #optimization algorithm to use to find the minimum loss/convergence
              optimizer= RMSprop(),
              #what metrics to use to test our validation data
              #Mean Absolute Value
              #Instead of evaluating the squared error, we'll use the absolute value
              #to be able to check the difference between the actual and our prediction
              metrics=["mae"])

#display the summary of our model
model.summary()
```

Let's understand our model summary.

The first layer shows a total Param of 640. This is the product of the numer of input neurons multipled by the output neurons.

In our model, we defined the input layer as having 9 input neurons(cylinders, displacement, horsepower, weight, acceleration, model year, origin_1, origin_2, origin_3) and 64 output neurons. If we multiple this, it is only 576. How did it become 640?

Because there is one neuron added as *bias* in the input of each layer. So 9 + 1 = 10 multiplied by 64 is equal to 640.

Same is true in the second layer which has 4,160 params. The input coming from the first layer is 64 + 1 bias = 65 multiplied by 64 = 4,160.

The last layer has 64 neurons + 1 bias = 65 multiplied by 1 output neuron = 65.

<div class="alert alert-block alert-info">
<b>Train the model.</b>
</div>

```python
#create a checkpoint for saving our model
#this way, we can save the best model throughout our training for future use
checkpoint_name = "ModelMPG-{epoch:03d}--{val_loss:.5f}.hdf5"
checkpoint = ModelCheckpoint(checkpoint_name, monitor="mean_absolute_error", verbose=1, save_best_only=True, mode="auto")

#create another callback to stop the training when it is not improving anymore
#patience is specify how many epochs to check if our monitor is improving
#we specify to monitor the validation loss
early_stop = EarlyStopping(monitor="val_loss", patience=10)

#traing our model and save the history of the iterations
history = model.fit(
                    #features of our training data
                    #in this example, it would be the "size" and "room"
                    x_train,
                    #the label/target that we are trying to predict
                    # that would be "price" in this example
                    y_train,
    
                    callbacks = [checkpoint, early_stop],#save our model f
                    #a good practice is to create a validation data from the training data
                    #if we don't have a separate data for validation
                    #as noted earlier, we should not validate on the test data
                    #0.2 means that we will assign 10% of the training data as validation data
                    #you can set a higher percentage if we have more training data
                    #during the training, the validation data is NEVER used in the training
                    validation_split=0.2,
    
                    #the number of iterations over the entire training data
                    #this is hyperparameter and will check how to adjust it later
                    epochs=800,
    
                    #the number of samples in one epoch
                    #the higher the number the more accurate our model would be
                    #the higher the number the higher memory is used
                    #default size is 32
                    #since we have a very small traing dataset, we just process it as one batch
                    batch_size=250,
    
                    #verbosity of the training progress, set to 0 if you don't want to display anything
                    verbose=2)
```

<div class="alert alert-block alert-info">
<b>Plot the result of our training.</b>
</div>

```python
#let's plot the history of our training
#create a range of the number of epochs we used
#we can derive this from the length of history of our mean_absolute_error list
#take note that the list starts with an index 0 but we will start
#with 1 so, we add 1 to the length so as not to lose the last item in the list
epochs = range(1, len(history.history["loss"]) + 1)

#let's define a figure with a wider width
plt.figure(figsize=(16,8))

#plot the training error
plt.plot(epochs, history.history["mean_absolute_error"], "b", label="Training MAE")
#together with validation error
plt.plot(epochs, history.history["val_mean_absolute_error"],"r", label="Validation MAE")

#write some labels
plt.title("History")
plt.xlabel = "Epochs"
plt.ylabel = "Error"

#display the legend
plt.legend()

#show the plot
plt.show()
```

By plotting the history, we can visualize the progress of our training. We can derive from the graph that at approximately at $670^{th}$ epoch, the error has flattened or converged.

<div class="alert alert-block alert-info">
<b>Test our data.</b>
</div>

```python
#evaluate the the data
y_predict = model.predict(x_test)

#compute and print the mean absolute error
loss = metrics.mean_absolute_error(y_test, y_predict)
print("Accuracy : +/- {0:.2f} MPG".format(loss))
```
